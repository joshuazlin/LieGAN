```python
import numpy as np
import matplotlib.pyplot as plt
import torch
import torch.nn as nn
import torch.nn.parallel
import torch.optim as optim
import math
from IPython.display import clear_output
from torch.nn import functional as F
```

```python
def w(x,n,a,b):
    return sum([a**m*np.cos(b**m*np.pi*x) for m in range(n)])

def s(x,n,h):
    b = np.linspace(-1,1,2*n)
    return np.interp(x,b,((-1)**np.array(range(2*n)))*h)
```

```python
"""
Test to show the Weierstrass function
"""
plt.figure(figsize=(8,8))
a = 0.5; b= 3; n = 100;
plt.plot([w(x,n,a,b) for x in np.linspace(-1,1,20000)],label=str(b))
plt.legend()
plt.show()
```

```python
"""
The actual WGAN/S^1 learning thingo
"""

class full_dense(nn.Module):
    def __init__(self,shape):
        """
        Shape looks something like 
        (1,n_1,n_2,...,1)
        """
        super(full_dense,self).__init__()
        main = nn.Sequential()
        for i in range(len(shape)-1):
            main.add_module(str(i)+":"+str(shape[i])+"->"+str(shape[i+1]),
                            nn.Linear(shape[i],shape[i+1]))
            if i != len(shape)-2:
                main.add_module(str(i)+":"+"relu",nn.LeakyReLU(0.2,inplace=True))
        self.main = main
        
    def forward(self, input):
        #print("forward")
        if isinstance(input.data, torch.cuda.FloatTensor) and self.ngpu > 1:
            output = nn.parallel.data_parallel(self.main, input, range(self.ngpu))
        else: 
            output = self.main(input)
         
        return output
    
class Linear_S1(nn.Linear):
    def __init__(self, in_features, out_features, bias=True,clip=None):
        super(Linear_S1,self).__init__(in_features,out_features,bias)
        self.clip = clip
    
    def forward(self, input):
        return F.linear(input, self.clip*torch.sin(self.weight), self.bias)

class full_dense_S1(nn.Module):
    def __init__(self,shape,clip):
        """
        Shape looks something like 
        (1,n_1,n_2,...,1)
        """
        super(full_dense_S1,self).__init__()
        main = nn.Sequential()
        for i in range(len(shape)-1):
            main.add_module(str(i)+":"+str(shape[i])+"->"+str(shape[i+1]),
                            Linear_S1(shape[i],shape[i+1],clip=clip))
            if i != len(shape)-2:
                main.add_module(str(i)+":"+"relu",nn.LeakyReLU(0.2,inplace=True))
        self.main = main
        
    def forward(self, input):
        #print("forward")
        if isinstance(input.data, torch.cuda.FloatTensor) and self.ngpu > 1:
            output = nn.parallel.data_parallel(self.main, input, range(self.ngpu))
        else: 
            output = self.main(input)
         
        return output
        
```

```python
settings = {
    "batchSize" : 64,
    "learningRate" : 0.001,
    "iters" : 5000,
    "log_rate" : 1000,
    "clip" : 0.5,
    "save" : False,
    "save_rate" : 50,
}
```

```python
settings_sweep = {
    "lrs" : np.linspace(0.0001,0.001,5),
    "repeat" : 20,
}

loss = []
loss_S1 = []
```

```python
for lr in settings_sweep["lrs"]:
    
    cum_loss = []
    cum_loss_S1 = []
    
    for i in range(settings_sweep["repeat"]):
    
        Dnet = full_dense(shape=(1,64,64,64,1))
        optimizerD = optim.RMSprop(Dnet.parameters(), lr = lr)
        noise = torch.FloatTensor((settings["batchSize"],1))

        Dnet_S1 = full_dense_S1(shape=(1,64,64,64,1),clip=settings["clip"])
        optimizerD_S1 = optim.RMSprop(Dnet_S1.parameters(), lr = lr)

        n = 6
        h = 1

        b0 = np.linspace(-1,1,300)
        b1 = torch.as_tensor(b0,dtype=torch.float32).resize_((300,1))

        cur_loss = []
        cur_loss_S1 = []

        for i in range(settings["iters"]):

            #data = np.random.random(size=settings["batchSize"])
            noise.resize_((settings["batchSize"],1)).uniform_(-1, 1)

            now = torch.as_tensor(s(noise,n,h),dtype=torch.float32)
            err = torch.sum(torch.abs(Dnet(noise)-now))
            err.backward()
            optimizerD.step()

            cur_loss.append(float((err).data.numpy()))

            err = torch.sum(torch.abs(Dnet_S1(noise)-now))
            err.backward()
            optimizerD_S1.step()

            cur_loss_S1.append(float((err).data.numpy()))

            for p in Dnet.parameters():
                p.data.clamp_(-settings["clip"], settings["clip"])
    
        #print(len(cur_loss))
        #print(len(cur_loss_S1))
        
        cum_loss.append(np.array(cur_loss))
        cum_loss_S1.append(np.array(cur_loss_S1))
    
    loss.append(sum(cum_loss)/settings_sweep["repeat"])
    loss_S1.append(sum(cum_loss_S1)/settings_sweep["repeat"])
```

```python
plt.figure(figsize=(8,8))
#for x in loss:
#    plt.plot(x,color='red')
#for y in loss_S1:
#    plt.plot(y,color='blue')

plt.gca().fill_between(range(len(loss[0])),
                       [min([loss[i][j] for i in range(len(loss))]) for j in range(len(loss[0]))],
                       [max([loss[i][j] for i in range(len(loss))]) for j in range(len(loss[0]))],
                      color='red',label='D',alpha=0.25)
plt.gca().fill_between(range(len(loss_S1[0])),
                       [min([loss_S1[i][j] for i in range(len(loss))]) for j in range(len(loss[0]))],
                       [max([loss_S1[i][j] for i in range(len(loss))]) for j in range(len(loss[0]))],
                      color='blue',label='D_S1',alpha=0.25)
plt.legend()
plt.show()
```

```python
Dnet = full_dense(shape=(1,64,64,64,1))
optimizerD = optim.RMSprop(Dnet.parameters(), lr = 0.5*settings["learningRate"])
noise = torch.FloatTensor((settings["batchSize"],1))
```

```python
Dnet_S1 = full_dense_S1(shape=(1,64,64,64,1),clip=settings["clip"])
optimizerD_S1 = optim.RMSprop(Dnet_S1.parameters(), lr = 0.25*settings["learningRate"])
```

```python
n = 6
h = 1

b0 = np.linspace(-1,1,300)
b1 = torch.as_tensor(b0,dtype=torch.float32).resize_((300,1))

if settings["save"] == True:
    data_save = []
    data_S1_save = []

for i in range(settings["iters"]):
    
    #data = np.random.random(size=settings["batchSize"])
    noise.resize_((settings["batchSize"],1)).uniform_(-1, 1)
    
    now = torch.as_tensor(s(noise,n,h),dtype=torch.float32)
    err = torch.sum(torch.abs(Dnet(noise)-now))
    err.backward()
    optimizerD.step()
    
    err = torch.sum(torch.abs(Dnet_S1(noise)-now))
    err.backward()
    optimizerD_S1.step()
    
    for p in Dnet.parameters():
        p.data.clamp_(-settings["clip"], settings["clip"])
        
    if settings["save"] == True:
        if i%settings["save_rate"] == 0:
            data_save.append(Dnet(b1).detach().numpy())
            data_S1_save.append(Dnet_S1(b1).detach().numpy())
    
    if i%settings["log_rate"] == 0:
        clear_output(wait=True)
        print(i)
        b = np.linspace(-1,1,2*n)
        plt.figure(figsize=(8,8))
        plt.plot(b,((-1)**np.array(range(2*n)))*h,label='truth')
        plt.plot(b0,Dnet(b1).detach().numpy(),label='D')
        plt.plot(b0,Dnet_S1(b1).detach().numpy(),label='D_S1')
        plt.ylim(-2*h,2*h)
        plt.legend()
        plt.show()
        
    
    
```
